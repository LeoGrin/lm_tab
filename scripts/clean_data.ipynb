```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/geonames-all-cities-with-a-population-1000@public.csv", sep=";")
df["target"] = df["Population"]
df.drop(["Geoname ID", "ASCII Name", "Alternate Names", "Country Code 2", "Admin1 Code", "Admin2 Code", "Admin3 Code", "Admin4 Code", "Population", "Modification date", "LABEL EN"], axis=1, inplace=True)
# split coordinates to latitude and longitude
df["latitude"] = df["Coordinates"].apply(lambda x: x.split(",")[0])
df["longitude"] = df["Coordinates"].apply(lambda x: x.split(",")[1])
df.drop(["Coordinates"], axis=1, inplace=True)
# only european cities, i.e "europe" in Timezone
df = df[df["Timezone"].apply(lambda x: "europe" in x.lower())]
# keep cities with population > 10_000
df = df[df["target"] > 10_000]
# log transform population
df["target"] = np.log(df["target"])
```

```python
df.to_parquet("../data/" + "european_cities_pop_log.parquet", index=False)
```

```python
df = pd.read_csv("../data/geonames-all-cities-with-a-population-1000@public.csv", sep=";")
df["target"] = df["Population"]
df.drop(["Geoname ID", "ASCII Name", "Alternate Names", "Country Code 2", "Admin1 Code", "Admin2 Code", "Admin3 Code", "Admin4 Code", "Population", "Modification date", "LABEL EN"], axis=1, inplace=True)
# split coordinates to latitude and longitude
df["latitude"] = df["Coordinates"].apply(lambda x: x.split(",")[0])
df["longitude"] = df["Coordinates"].apply(lambda x: x.split(",")[1])
df.drop(["Coordinates"], axis=1, inplace=True)
# only keep cities with population > 10_000
df = df[df["target"] > 10_000]
# log transform population
df["target"] = np.log(df["target"])
# remove Timezone with less than 1000 cities
df = df.groupby("Timezone").filter(lambda x: len(x) > 1000)
```

```python
df.to_parquet("../data/" + "all_cities_pop_log.parquet", index=False)
```

```python
import pandas as pd
df = pd.read_csv("../data/Movies_dataset.csv")
```

```python
df
```

```python
import numpy as np
# convert budget to numerical
df['Revenue'] = df['Revenue'].replace('-', 'NaN')
df['Revenue'] = df['Revenue'].replace('[\$,]' , '', regex=True)
df['Revenue'] = pd.to_numeric(df['Revenue'], errors='coerce')
# log transform
df['Revenue'] = np.log(df['Revenue'])
```

```python
# convert Release_Date to datetime
df['Release_Date'] = pd.to_datetime(df['Release_Date'])
```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# Assuming df is your DataFrame
# df = pd.read_csv('your_data.csv')

# Convert 'Release_Date' to datetime format
df['Release_Date'] = pd.to_datetime(df['Release_Date'])

# Extract the release year from 'Release_Date' and create a new column 'Release_Year'
df['Release_Year'] = df['Release_Date'].dt.year

# Convert 'Revenue' to numeric format
df['Revenue'] = pd.to_numeric(df['Revenue'], errors='coerce')

# Define a MinMaxScaler
scaler = MinMaxScaler()

# Apply the MinMaxScaler to the 'Revenue' column for each year
df['Revenue_Normalized'] = df.groupby('Release_Year')['Revenue'].transform(lambda x: scaler.fit_transform(x.values.reshape(-1, 1)).flatten())

# To view the dataframe with the new column
df
```

```python
df = df[["Movie_Name", "Certification", "Release_Date", "Genres", "Language", "Revenue_Normalized", "Runtime"]]
```

```python
# drop nans in Revenue
df = df.dropna(subset=['Revenue_Normalized'])
```

```python
# rename Movie_Name to name
df = df.rename(columns={"Movie_Name": "name", "Revenue_Normalized":"target"})
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
# export to parquet
df.to_parquet("../data/movies_2023_normalized.parquet")
```

```python
from bs4 import BeautifulSoup
import requests

url = "https://huggingface.co/spaces/mteb/leaderboard"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

table = soup.find('table') # find the table in the HTML
print(table)
rows = table.find_all('tr') # find all the table rows

for row in rows:
    cols = row.find_all('td') # find all columns in each row
    cols = [col.text.strip() for col in cols] # clean up the column content
    print(cols)  # print the columns
```
