```python
from skrub import TableVectorizer
def return_features(X, encoder_name, dataset_name=None, use_cache=True, override_cache=False, cardinality_threshold=30, fail_if_not_cached=False):
    tb = TableVectorizer(cardinality_threshold=cardinality_threshold,
                        high_card_cat_transformer = "passthrough",
                        low_card_cat_transformer = "passthrough",
                        numerical_transformer = "passthrough",
                        datetime_transformer = "passthrough",
    ) #just to get the high cardinality columns
    tb.fit(X)
    res = [(name, cols) for (name, _, cols) in tb.transformers_]
    return res
```

```python
import pandas as pd
from src.data_loading import load_data

# Initialize an empty DataFrame to store the features
features_df = pd.DataFrame()

for dataset_name in datasets:
    # Load the data
    X, y = load_data(dataset_name, max_rows=10_000)

    # Get the high cardinality and other features
    tuples = return_features(X, 'encoder_name', dataset_name)

    res = {'dataset': dataset_name}
    for (name, col) in tuples:
        # remove _transformer suffix
        name_ = name[-len("_transformer"):]
        if name_ == "high_card_cat":
            name_ =  "text"
        res[name] = ", ".join(col)
    print(res)

    # Create a temporary DataFrame to store the features for this dataset
    for key in res:
        res[key] = [res[key]] if not isinstance(res[key], list) else res[key]

    temp_df = pd.DataFrame(res)

    # Append the temporary DataFrame to the main DataFrame
    features_df = pd.concat([features_df, temp_df], ignore_index=True)


    print(features_df)
```

```python
features_df.drop("remainder", inplace=True, axis=1)
```

```python
def truncate_long_entries(data, max_length=40):
    """Truncate entries in DataFrame that exceed max_length."""
    for col in data.columns:
        data[col] = data[col].apply(lambda x: (x[:max_length] + '...') if len(str(x)) > max_length else x)
    return data
```
